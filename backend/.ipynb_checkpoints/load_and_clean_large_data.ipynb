```python
import unicodedata

def are_equal_ignore_accents(str1, str2):
    # Normalize the strings to remove accents
    def normalize_and_map(s):
        normalized = unicodedata.normalize('NFKD', s)
        mapped = normalized.replace("ø", "o").replace("æ", "ae").replace("Ø", "O")
        ascii_str = mapped.encode('ascii', 'ignore').decode('utf-8')
        return ascii_str
    
    return normalize_and_map(str1) == normalize_and_map(str2)
are_equal_ignore_accents("æ", "ae")
```

```python
def get_id_for_player(player_name, id_response_json):
    player_id = ""
    if "results" in id_response_json and id_response_json["results"]:
        for res in id_response_json['results']:
            if are_equal_ignore_accents(player_name, res["name"]):
                player_id=res["id"]
                break
    return player_id
```

```python
def getAge(dt):
    present = datetime.now()
    year_diff = present.year-dt.year
    if dt.month > present.month or (dt.month==present.month and dt.day > present.day): year_diff-=1
    return year_diff
```

```python
with open("player_id.json") as f: queries = json.load(f)
players_df = pd.DataFrame(columns=["id", "player_name"])
bad_names = []
prevName = ""
players_json_index = 0
for query in queries:
    player_name = query["query"]
    player_id, player_ids = "", []
    num_matching_names = 0
    for res in query['results']:
        if are_equal_ignore_accents(player_name, res["name"]):
            player_ids.append([res["id"], res["age"]])
            num_matching_names+=1
    if not player_ids: 
        bad_names.append(player_name)
        continue
    if num_matching_names>=2:
        if prevName == player_name: players_json_index+=1
        else: players_json_index = find_index_of_player(player_name)
        if "b" in players_json[players_json_index]: 
            birth_date = datetime.strptime(players_json[players_json_index]["b"], "%d/%m/%Y")
            for i in player_ids:
                if i[1]!="-" and getAge(birth_date)==int(i[1]):
                    player_id = i[0]
            prevName = player_name
            player_name = f"{player_name} ({birth_date.year})"
        else: player_id = player_ids[0][0]
    else:
        num_matching_names=0
        player_id = player_ids[0][0]
        prevName = ""
    if player_id: players_df = pd.concat([players_df, pd.DataFrame([{"id": int(player_id), "player_name": player_name}])], ignore_index=True)
    else: bad_names.append(player_name)
```

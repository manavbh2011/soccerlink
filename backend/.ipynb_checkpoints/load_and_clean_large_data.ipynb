```python
import requests, csv, numpy as np, os, json
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
```

```python
with open("players_large.json") as f:
    data = json.loads(f.read())
```

```python
import unicodedata

def are_equal_ignore_accents(str1, str2):
    # Normalize the strings to remove accents
    normalized_str1 = unicodedata.normalize('NFKD', str1).encode('ascii', 'ignore').decode('utf-8')
    normalized_str2 = unicodedata.normalize('NFKD', str2).encode('ascii', 'ignore').decode('utf-8')
    
    # Compare the normalized strings
    return normalized_str1 == normalized_str2
are_equal_ignore_accents("resume", "resum√©")
```

```python
def get_id_for_player(player_name, id_response_json):
    player_id = ""
    if "results" in id_response_json and id_response_json["results"]:
        for res in id_response_json['results']:
            if are_equal_ignore_accents(player_name, res["name"]):
                player_id=res["id"]
                break
    return player_id
```

```python
import requests, csv, numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://ainsworthsports.com/soccer_player_rankings_all_time_1_to_1000.htm"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
tags = soup.find_all("td")
df = pd.DataFrame()
players = pd.DataFrame(columns={"id", "player_name"})
bad_names = []
for i in range(29, len(tags), 4): #jump to first name tag, names are every 4 tags
    player_name = tags[i].text.strip()
    id_response_json = requests.get(f"http://localhost:8000/players/search/{player_name}").json()
    if "results" in id_response_json and id_response_json["results"]:
        print(player_name)
        player_id = id_response_json['results'][0]["id"]
        teams_response = requests.get(f"http://localhost:8000/players/{player_id}/transfers")
        players = pd.concat([players, pd.DataFrame([{"id": int(player_id), "player_name": player_name}])], ignore_index=True)
        df = pd.concat([df, pd.DataFrame([teams_response.json()])], ignore_index=True)
    else:
        bad_names.append(player_name)
df
```

```python
def add_player_to_df(player_name):
    global players_df
    id_response_json = requests.get(f"http://localhost:8000/players/search/{player_name}").json()
    if id_response_json["results"]:
        player_id = id_response_json['results'][0]["id"]
        teams_response = requests.get(f"http://localhost:8000/players/{player_id}/transfers")
        players_df = pd.concat([pd.DataFrame([teams_response.json()]), players_df], ignore_index=True)
    else:
        print(f"Failed for {player_name}")
add_player_to_df("Maicon")
add_player_to_df("Heung-Min Son")
add_player_to_df("Ji-Sung Park")
players_df
```

```python
transfers = players_df[['transfers']]
formatted_df = pd.DataFrame(columns=["player_id", "team_name", "start_date", "end_date"])
for i, row in transfers.iterrows():
    row = row[0]
    for j, transfer in enumerate(row):
        new_row_dict = {}
        if transfer["to"]["clubName"] not in ["Retired", "Without Club", "Career break"]: 
            new_row_dict["player_id"] = int(players_df.iloc[i]["id"])
            new_row_dict["team_name"] = transfer["to"]["clubName"]
            new_row_dict["start_date"] = transfer["date"]
            if j>0: new_row_dict["end_date"] = row[j-1]["date"]
            else: new_row_dict["end_date"] = np.nan
            formatted_df = pd.concat([formatted_df, pd.DataFrame([new_row_dict])], ignore_index=True)
formatted_df
```

```python
from difflib import SequenceMatcher
def similarity(str1, str2):
    return SequenceMatcher(None, str1, str2).ratio()
```

```python
counts = formatted_df["team_name"].value_counts()
for name, count in counts.iteritems():
    closest_team, max_sim = "", 0
    for team in teams_list["team_name"]:
        sim = similarity(name, team)
        if sim>max_sim: 
            max_sim = sim
            closest_team = team
    print(name, closest_team)
```

```python
formatted_df_cleaned_ids = formatted_df.copy()
counts = formatted_df["team_name"].value_counts().rank(method="first", ascending=False).astype(int)
formatted_df_cleaned_ids["team_name"] = formatted_df_cleaned_ids["team_name"].map(counts)
formatted_df_cleaned_ids.rename(columns={'team_name': 'team_id'}, inplace=True)
formatted_df_cleaned_ids
```

```python
formatted_df_cleaned_ids["start_date"] = pd.to_datetime(formatted_df_cleaned_ids["start_date"]).dt.strftime('%m/%d/%Y')
formatted_df_cleaned_ids["end_date"] = pd.to_datetime(formatted_df_cleaned_ids["end_date"]).dt.strftime('%m/%d/%Y')
```

```python
formatted_df_cleaned_ids.to_csv('players_and_teams_medium.csv', index=True, index_label="id")
```
